```python
# append labels to canon dataframe from manually sorting cropped pictures
import os
import pandas as pd
import numpy as np
from os.path import join as pj
pd.set_option("display.notebook_repr_html", False)
```

```python
label_dir = "../data/interim/170308/labelled/"
dirs = ["good_sign", "poor_sign", "not_sign"]
labels = ["good", "poor", "not"]
label_dic = {}
for d, l in zip(dirs, labels):
    for filename in os.listdir(pj(label_dir, d)):
        if filename.endswith("jpg"):
            circle_id = filename[:-4]
            label_dic[circle_id] = l
label_dic
```

```python
df = pd.read_csv("../data/interim/170308/dataframes/canon_1837_19295.csv")
df.set_index("id", inplace=True)
df["is_sign"] = pd.Series(label_dic)
df.head()
```

```python
np.unique(df["is_sign"]) 
```

```python
#remove those rows for which there are no labels
df.dropna(subset=["is_sign"], inplace=True)
df.head()
```

```python
df.to_csv("../data/interim/170308/dataframes/canon_1837_19295_labelled.csv")
```
